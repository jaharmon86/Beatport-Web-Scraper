```python
import requests
from bs4 import BeautifulSoup
resp = requests.get("https://www.beatport.com/genre/progressive-house/15/top-100")
soup = BeautifulSoup(resp.text, "html.parser")
result_rows = []
import pandas as pd
lis = soup.find_all("li", {"class": "bucket-item ec-item track"})
for l in lis[0:]:
    # need all a's with attribute data-artist
    artists = l.find_all('a')
    found_artists = []
    for x in artists:
        if "data-artist" in x.attrs:
            found_artists.append(x.get_text())
    title = l.find('span', {"class": "buk-track-primary-title"}).get_text()
    date = l.find('p', {"class": "buk-track-released"}).get_text()
    result_rows.append({
        "artists": found_artists,
        "track_title": title,
        "release": date
    })
df = pd.DataFrame.from_records(result_rows)
df = df[['artists', 'track_title', 'release']]
#df['artists'] = df['artists'].str.get(0)
df['artists'] = df['artists'].str.join(', ')
df
```
